### 3. Lưu dữ liệu vào file csv

**3.1 Dữ liệu lịch sử giá theo ngày trong 5 năm gần nhất**

```python
import pandas as pd
from vnstock3 import Vnstock

def fetch_and_save_data_per_year(start_year, end_year):
    # Khởi tạo đối tượng Vnstock
    stock_api = Vnstock()

    # Lấy danh sách mã cổ phiếu niêm yết trên sàn HOSE
    hose_symbols = stock_api.stock(symbol='A32', source='VCI').listing.symbols_by_group('HOSE')
    
    # Tạo danh sách mã cổ phiếu từ kết quả Series
    symbols = hose_symbols.tolist()

    # Lặp qua các năm để lấy và lưu dữ liệu
    for year in range(start_year, end_year + 1):
        print(f"Đang lấy dữ liệu cho năm {year}...")
        all_price_data = []
        
        for symbol in symbols:
            try:
                # Khởi tạo đối tượng Vnstock cho mã cổ phiếu hiện tại
                stock = Vnstock().stock(symbol=symbol, source='VCI')
                
                # Lấy dữ liệu giá cổ phiếu theo năm
                price_data = stock.quote.history(start=f'{year}-01-01', end=f'{year}-12-31', interval='1D')
                
                # Thêm mã cổ phiếu vào dữ liệu
                price_data['symbol'] = symbol
                
                # Thêm dữ liệu vào danh sách
                all_price_data.append(price_data)
                
                print(f"Đã lấy dữ liệu cho mã cổ phiếu: {symbol}")
            except Exception as e:
                print(f"Lỗi khi lấy dữ liệu cho mã cổ phiếu {symbol}: {e}")
        
        # Kết hợp tất cả dữ liệu vào một DataFrame
        if all_price_data:
            all_price_data_df = pd.concat(all_price_data, ignore_index=True)
            
            # Lưu dữ liệu vào file CSV theo năm
            all_price_data_df.to_csv(f'all_stock_data_{year}.csv', index=False)
            print(f"Dữ liệu đã được lưu vào file all_stock_data_{year}.csv")
        else:
            print(f"Không có dữ liệu cho năm {year}")

# Thực hiện việc lưu dữ liệu từ năm 2019 đến năm 2024
fetch_and_save_data_per_year(2021, 2024)
```

```python
import pandas as pd
from vnstock3 import Vnstock
from datetime import datetime, timedelta

def fetch_and_save_hourly_data_from_july():
    # Khởi tạo đối tượng Vnstock
    stock_api = Vnstock()

    # Lấy danh sách mã cổ phiếu niêm yết trên sàn HOSE
    hose_symbols = stock_api.stock(symbol='A32', source='VCI').listing.symbols_by_group('HOSE')
    
    # Tạo danh sách mã cổ phiếu từ kết quả Series
    symbols = hose_symbols.tolist()

    # Xác định năm và tháng hiện tại
    today = datetime.today()
    current_year = today.year
    current_month = today.month

    # Bắt đầu từ tháng 7 của năm hiện tại
    start_year = current_year
    start_month = 7

    # Lặp qua các tháng từ tháng 7 năm hiện tại đến tháng hiện tại
    for year in range(start_year, current_year + 1):
        for month in range(start_month if year == start_year else 1, current_month + 1 if year == current_year else 13):
            # Tính toán ngày bắt đầu và ngày kết thúc của tháng
            start_date_str = f'{year}-{month:02d}-01'
            if month == 12:
                end_date_str = f'{year+1}-01-01'
            else:
                end_date_str = f'{year}-{month+1:02d}-01'

            print(f"Đang lấy dữ liệu theo giờ cho tháng {month:02d}/{year}...")
            all_hourly_data = []

            for symbol in symbols:
                try:
                    # Khởi tạo đối tượng Vnstock cho mã cổ phiếu hiện tại
                    stock = Vnstock().stock(symbol=symbol, source='VCI')

                    # Lấy dữ liệu giá cổ phiếu theo giờ
                    hourly_data = stock.quote.history(start=start_date_str, end=end_date_str, interval='1H')

                    # Thêm mã cổ phiếu vào dữ liệu
                    hourly_data['symbol'] = symbol

                    # Thêm dữ liệu vào danh sách
                    all_hourly_data.append(hourly_data)

                    print(f"Đã lấy dữ liệu theo giờ cho mã cổ phiếu: {symbol}")
                except Exception as e:
                    print(f"Lỗi khi lấy dữ liệu theo giờ cho mã cổ phiếu {symbol}: {e}")

            # Kết hợp tất cả dữ liệu vào một DataFrame
            if all_hourly_data:
                all_hourly_data_df = pd.concat(all_hourly_data, ignore_index=True)

                # Lưu dữ liệu vào file CSV theo tháng
                all_hourly_data_df.to_csv(f'hourly_stock_data_{year}_{month:02d}.csv', index=False)
                print(f"Dữ liệu đã được lưu vào file hourly_stock_data_{year}_{month:02d}.csv")
            else:
                print(f"Không có dữ liệu cho tháng {month:02d}/{year}")

# Lưu dữ liệu theo giờ từ tháng 7 năm hiện tại đến tháng hiện tại
fetch_and_save_hourly_data_from_july()
```

```python
import pandas as pd
from vnstock3 import Vnstock
from datetime import datetime, timedelta

def fetch_and_save_minute_data(start_index, end_index, file_suffix):
    # Khởi tạo đối tượng Vnstock
    stock_api = Vnstock()

    # Lấy danh sách mã cổ phiếu niêm yết trên sàn HOSE
    hose_symbols = stock_api.stock(symbol='A32', source='VCI').listing.symbols_by_group('HOSE')
    
    # Tạo danh sách mã cổ phiếu từ kết quả Series
    symbols = hose_symbols.tolist()

    # Tính toán ngày bắt đầu và ngày kết thúc
    end_date = datetime.today()
    start_date = end_date - timedelta(days=7)
    start_date_str = start_date.strftime('%Y-%m-%d')
    end_date_str = end_date.strftime('%Y-%m-%d')

    # Lấy dữ liệu cho các mã chứng khoán từ start_index đến end_index
    all_minute_data = []
    for index in range(start_index, min(end_index, len(symbols))):
        symbol = symbols[index]
        try:
            # Khởi tạo đối tượng Vnstock cho mã cổ phiếu hiện tại
            stock = Vnstock().stock(symbol=symbol, source='VCI')
            
            # Lấy dữ liệu giá cổ phiếu theo phút
            minute_data = stock.quote.history(start=start_date_str, end=end_date_str, interval='1m')
            
            # Thêm mã cổ phiếu vào dữ liệu
            minute_data['symbol'] = symbol
            
            # Thêm dữ liệu vào danh sách
            all_minute_data.append(minute_data)
            
            print(f"Đã lấy dữ liệu theo phút cho mã cổ phiếu: {symbol}")
        except Exception as e:
            print(f"Lỗi khi lấy dữ liệu theo phút cho mã cổ phiếu {symbol}: {e}")

    # Kết hợp tất cả dữ liệu vào một DataFrame
    all_minute_data_df = pd.concat(all_minute_data, ignore_index=True)

    # Lưu dữ liệu vào file CSV
    all_minute_data_df.to_csv(f'minute_stock_data_part_{file_suffix}.csv', index=False)

    print(f"Dữ liệu đã được lưu vào file minute_stock_data_part_{file_suffix}.csv")

# Lưu dữ liệu theo phút của 100 mã chứng khoán đầu tiên trên sàn HOSE
fetch_and_save_minute_data(300, 450, 4)
```

```python
import pandas as pd
from vnstock3 import Vnstock

def fetch_and_save_data(start_year, end_year):
    # Khởi tạo đối tượng Vnstock
    stock_api = Vnstock()

    # Lấy danh sách mã cổ phiếu niêm yết trên sàn HOSE
    hose_symbols = stock_api.stock(symbol='A32', source='VCI').listing.symbols_by_group('HOSE')
    
    # Tạo danh sách mã cổ phiếu từ kết quả Series
    symbols = hose_symbols.tolist()

    # Khởi tạo danh sách để lưu tất cả dữ liệu
    all_price_data = []

    # Lặp qua các năm để lấy và lưu dữ liệu
    for year in range(start_year, end_year + 1):
        print(f"Đang lấy dữ liệu cho năm {year}...")
        
        for symbol in symbols:
            try:
                # Khởi tạo đối tượng Vnstock cho mã cổ phiếu hiện tại
                stock = Vnstock().stock(symbol=symbol, source='VCI')
                
                # Lấy dữ liệu giá cổ phiếu theo năm
                price_data = stock.quote.history(start=f'{year}-01-01', end=f'{year}-12-31', interval='1D')
                
                # Thêm mã cổ phiếu vào dữ liệu
                price_data['symbol'] = symbol
                
                # Thêm dữ liệu vào danh sách
                all_price_data.append(price_data)
                
                print(f"Đã lấy dữ liệu cho mã cổ phiếu: {symbol}")
            except Exception as e:
                print(f"Lỗi khi lấy dữ liệu cho mã cổ phiếu {symbol}: {e}")
    
    # Kết hợp tất cả dữ liệu vào một DataFrame
    if all_price_data:
        all_price_data_df = pd.concat(all_price_data, ignore_index=True)
        
        # Lưu dữ liệu vào file CSV duy nhất
        all_price_data_df.to_csv('all_stock_data_2021_to_present.csv', index=False)
        print(f"Dữ liệu đã được lưu vào file all_stock_data_2021_to_present.csv")
    else:
        print("Không có dữ liệu nào được lưu.")

# Thực hiện việc lưu dữ liệu từ năm 2021 đến hiện tại
fetch_and_save_data(2021, 2024)
```
